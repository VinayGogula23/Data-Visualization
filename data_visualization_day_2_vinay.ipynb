##### 1. Write a Python program to display a bar chart of the fare of class.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data=pd.read_csv(r"C:\Users\ADMIN\python files\assignments\Data_visualization\Data_visualization_day_2\titanic.csv")
data
```

```python
x=data[data['class']=='First']['fare'].mean()
y=data[data['class']=='Second']['fare'].mean()
z=data[data['class']=='Third']['fare'].mean()
plt.bar(['Frist','Second','Thrid'],[x,y,z])
plt.ylabel('mean of fare')
```

```python
x=data[data['sex']=='male']['fare'].mean()
y=data[data['sex']=='female']['fare'].mean()
plt.bar(['male','female'],[x,y])
plt.ylabel('mean of fare')
```

###### 2. Write a Python program to display a horizontal bar chart of the fare of class.

```python
plt.barh(['Frist','Second','Thrid'],[x,y,z])
```

###### 3. Write a Python program to display a bar chart of the fare of class. Use different color for each bar.

```python
x=data[data['class']=='First']['fare'].mean()
y=data[data['class']=='Second']['fare'].mean()
z=data[data['class']=='Third']['fare'].mean()
sb.barplot(['Frist','Second','Thrid'],[x,y,z])
plt.ylabel('mean of fare')
```

###### 4. Write a Python program to create a pie chart of the fare of class.

```python
plt.pie([x,y,z],labels=['Frist','Second','Thrid'],autopct='%0.1f%%',radius=2)
plt.show()
```
