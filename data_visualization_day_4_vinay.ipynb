#### Write a Python program to display all Univariate Plots from the dataframe.


1.   Numeric based plots
2.   Categorical based plots




```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data=pd.read_csv(r"C:\Users\Gogula Vinay\Desktop\Innomatics labs\Data Science\lms assignment\10. Data Visualization\Data_visualization\Data_visualization_day_4\diamonds.csv")
data.drop(columns='Unnamed: 0',inplace=True)
data
```

```python
plt.figure(figsize=(15,5))
sb.histplot(data=data,x='carat',bins=3,hue='cut') 
```

```python
plt.figure(figsize=(5,5))
sb.stripplot(data=data,y='depth',x='cut')
plt.xticks(rotation=90)
plt.show()
```

#### Write a Python program to display all Bivariate plots from the dataframe.

1.   Numeric to Numeric based plots
2.   Numeric to Categorical Plots
3.   Categorical to Categorical Plots






```python
plt.figure(figsize=(20,15))
sb.scatterplot(data=data,x ='carat', y = 'price',hue='cut')
```

```python
plt.figure(figsize=(20,15))
sb.lineplot(data=data,x ='carat', y = 'price',hue='cut')
```

```python
plt.figure(figsize=(50,15))
sb.violinplot(data=data,x ='carat', y = 'price',hue='cut')
plt.xticks(rotation=90)
plt.show()
```
