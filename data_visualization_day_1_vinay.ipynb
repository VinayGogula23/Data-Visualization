#### 1. create a bar plot for count the frequency of the three species.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import warnings 
warnings.filterwarnings('ignore')
import nltk
```

```python
data=pd.read_csv(r"C:\Users\Gogula Vinay\Desktop\Innomatics labs\Data Science\lms assignment\10. Data Visualization\Data_visualization\Data_visualization_day_1\Iris.csv")
data.drop(columns='Id',inplace=True)
data
```

```python
sb.countplot(x=data['Species'])
```

```python
x=data[data['Species']=='Iris-setosa']['SepalLengthCm'].count()
y=data[data['Species']=='Iris-versicolor']['SepalLengthCm'].count()
z=data[data['Species']=='Iris-virginica']['SepalLengthCm'].count()
plt.bar(['Iris-setosa','Iris-versicolor','Iris-virginica'],[x,y,z])
plt.ylabel('count')
plt.show()
```

##### 2. Create a pie chart to show the percentage of species.

```python
plt.pie([x,y,z],labels=['Iris-setosa','Iris-versicolor','Iris-virginica'],autopct='%0.1f%%',radius=2)
plt.show()
```

##### 3. Create the joint to all numberical variable for analyze the relationship between two variables and describe their individual distributions on the same plot.

```python
sb.jointplot(data=data,x='PetalLengthCm',y='PetalWidthCm',hue='Species')
plt.show()
# we differentiate setosa completely with pl variable on distribution plot but with pw variable there is a small amount of error,
#when we see distribution plot with respect to pl or pw variables we cant differentiate versicolor and virginica completely
```

##### 4. Create a scatter plot and take hue = "Species"

```python
sb.scatterplot(data=data,x='SepalLengthCm',y='SepalWidthCm',hue='Species')
plt.show()
```
